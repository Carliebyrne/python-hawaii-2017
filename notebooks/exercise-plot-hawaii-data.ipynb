```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('hawaii-temperature-data.txt', comments='%')
```

```python
year = data[:, 0]
```

```python
month = data[:, 1]
```

```python
temperature_anomaly = data[:, 2]
```

```python
time = year + 1/12*(month - 1)
```

```python
time
```

```python
plt.figure()
plt.plot(time, temperature_anomaly, '-k')
plt.plot(time, data[:, 8], '-r', linewidth=5)
plt.xlabel('time (years)')
plt.ylabel('temperature anomaly (ºC)')
plt.title('Temperature anomaly in Hawaii')
plt.savefig('temperature-anomaly-hawaii.jpg')
plt.show()
```
